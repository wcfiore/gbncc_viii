```python
import numpy as np
import matplotlib.pyplot as plt
from glob import glob
```

```python
datadir = "./data"
profiles = sorted(glob(f"{datadir}/*fiore+22.profile"))
```

```python
def plot_profile(fname):
    prof = np.loadtxt(fname,usecols=3,skiprows=1,dtype=float)
    nbins = len(prof)
    bins = np.arange(0,nbins,1)
    phase = [b/(nbins-1) for b in bins]
    psrname, freq, junk = fname.split("/")[-1].split("_")
    plt.plot(phase,prof)
    plt.xlim(0,1)
    plt.xlabel("Pulse Phase [rotations]")
    plt.ylabel("Flux [arb.]")
    plt.title(f"PSR {psrname} at {freq}")
    plt.show()
```

```python
for fname in profiles:
    plot_profile(fname)
```
